```python
# 2SUM
# 使用哈希表
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        t = {}
        for i, j in enumerate(nums):
            res = target - j
            if res in t:
                return [i, t[res]]
            else:
                t[j] = i
        
```

```python
## 15 TreeSum
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        
        nums = sorted(nums)
        target = 0
        final_res = []
        n = len(nums)
        last = None
        for i in range(n):
            if nums[i] == last:  #[ ]跳过重复 
                continue
            other = target - nums[i]
            tuples = self.twoSum(nums, i+1, other)
            for t in tuples:
                final_res.append((*t, nums[i]))
            last = nums[i] #[ ]跳过重复
        return final_res

    def twoSum(self, nums, start, target):
        res = []
        lo = start ##[ ]重点
        hi = len(nums)-1
        while lo < hi:
            left = nums[lo]
            right = nums[hi]
            s = left + right
            if s > target:
                while lo<hi and nums[hi] == right:
                    hi -= 1
            elif s < target:
                while lo < hi  and nums[lo] == left:
                    lo += 1
            else:
                res.append((left, right))  #[ ]注意不要写错，是返回元素不是索引
                while lo<hi and nums[hi] == right:
                    hi -= 1
                while lo<hi and nums[lo] == left:
                    lo += 1
        return res
```

```python
## nSum

## fourSum
class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        return self.nSum(nums,4,0,target)

    def nSum(self, nums, n, start, target):
        res = []
        size = len(nums)
        if n<2 or size<n:  #[ ]注意结束条件
            return res
        if n == 2:
            l, r = start, size-1
            while l < r:
                left = nums[l]
                right = nums[r]
                s = nums[l] + nums[r]
                if s < target:
                    while l<r and nums[l]==left:
                        l+=1
                elif s > target:
                    while l<r and nums[r]==right: 
                        r-=1
                else:
                    res.append([left,right])
                    while l<r and nums[l]==left:
                        l+=1
                    while l<r and nums[r]==right:
                        r-=1
        else:
            last = None  
            for i in range(start, size):
                if nums[i]==last:  #[ ]注意跳过条件
                    continue
                sub = self.nSum(nums, n-1, i+1, target-nums[i])
                for arr in sub:
                    if arr:
                        res.append([nums[i]]+arr)
                last = nums[i]
                
        return res
```
