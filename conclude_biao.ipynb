```python
# 21
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(-1)
        p = dummy
        p1 = list1
        p2 = list2
        
        while p1 and p2:
            if p1.val > p2.val:
                p.next = p2
                p2 = p2.next
            else:
                p.next = p1
                p1 = p1.next
            p = p.next
            
        if p1:
            p.next = p1
        if p2:
            p.next = p2
        return dummy.next
```

```python
# 86
class Solution:
    def partition(self, head: Optional[ListNode], x: int) -> Optional[ListNode]:
        dummy1 = ListNode(-1)
        dummy2 = ListNode(-1)
        p1 = dummy1
        p2 = dummy2
        p = head
        
        while p:
            if p.val < x:
                p1.next = p
                p1 = p1.next
                
            else:
                p2.next = p
                p2 = p2.next
            p = p.next
        p2.next = None
        p1.next = dummy2.next
        return dummy1.next
```

```python
#23
import heapq
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if not lists:
            return None
        dummy = ListNode(-1)
        p = dummy
        pq = []
        i = 0
        for head in lists:
            if head:
                heapq.heappush(pq, (head.val,i, head))
                i+=1
        while pq:
            node = heapq.heappop(pq)[2]
            p.next = node
            if node.next:
                heapq.heappush(pq, (node.next.val, i, node.next))
                i+=1
            p = p.next
        return dummy.next   
```

```python
# 19
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        #寻找倒数第k+1个节点
        dummy = ListNode(-1)
        dummy.next = head
        p = self.findFromEnd(dummy, n+1)
        p.next = p.next.next
        return dummy.next



    # 返回链表的倒数第 k 个节点
    def findFromEnd(self, head: ListNode, k: int) -> ListNode:
        p1 = head
        # p1 先走 k 步
        for i in range(k):
            p1 = p1.next
        p2 = head
        # p1 和 p2 同时走 n - k 步
        while p1 != None:
            p1 = p1.next
            p2 = p2.next
        # p2 现在指向第 n - k 个节点，也就是倒数第k个节点
        return p2
```

```python
# 160
class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        p1 = headA
        p2 = headB
        while p1 != p2:
            if not p1:
                p1 = headB
            else:
                p1 = p1.next

            if not p2:
                p2 = headA
            else:
                p2 = p2.next
        
        return p1
```

```python
# 92
class Solution:
    def reverseBetween(self, head: Optional[ListNode], left: int, right: int) -> Optional[ListNode]:

        
        successor = None # 后继节点
        # 反转以 head 为起点的 n 个节点，返回新的头节点
        def reverseN(head: ListNode, n: int) -> ListNode:
            global successor
            if n == 1:
                # 记录第 n + 1 个节点
                successor = head.next
                return head
            # 以 head.next 为起点，需要反转前 n - 1 个节点
            last = reverseN(head.next, n - 1)

            head.next.next = head
            # 让反转之后的 head 节点和后面的节点连起来
            head.next = successor
            return last 
        
        # 递归，对head翻转(left, right)也就是对head.next翻转(left-1, right-1)
        # 直到left-1==0的时候
        if left == 1:
            return reverseN(head, right)
        head.next = self.reverseBetween(head.next, left-1, right-1)
        return head
```

```python
# 26
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        
        slow, fast = 0,0
        while fast < len(nums):
            if nums[fast] != nums[slow]:
                slow += 1
                nums[slow] = nums[fast]
            fast += 1
        return slow+1 #元素数量是最大索引+1
```

```python
# 83
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head:
            return head
        fast, slow = head, head
        while fast is not None:
            if fast.val != slow.val:
                slow.next = fast
                slow = slow.next
            fast = fast.next
        slow.next = None
        return head
```

```python
# 167
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        left, right = 0, len(numbers)-1
        while left < right:
            s = numbers[left] + numbers[right]
            if s == target:
                return [left+1, right+1]
            elif s > target:
                right -= 1
            elif s < target:
                left += 1
```

```python
# 344
class Solution:
    def reverseString(self, s: List[str]) -> None:
        """
        Do not return anything, modify s in-place instead.
        """
        i, j = 0, len(s)-1
        while i < j:
            temp = s[i]
            s[i] = s[j]
            s[j] = temp

            i += 1
            j -= 1
```

```python
# 283
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        slow, fast = 0, 0
        while fast < len(nums):
            if nums[fast] != 0:
                nums[slow] = nums[fast]
                slow += 1
            fast += 1
        for i in range(slow, len(nums)):
            nums[i] = 0
```

```python
# 322 [x]
class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:

        dp = [amount +1 for i in range(amount+1)]
        dp[0] = 0
        for i in range(len(dp)):
            for coin in coins:
                if i < coin:
                    continue
                dp[i] = min(dp[i], dp[i-coin]+1)
        return dp[amount] if dp[amount]!=amount+1 else -1
```

```python
# 0-1 背包问题

def knapsack(W: int, N: int, wt: List[int], val: List[int]) -> int:
    assert N == len(wt)
    # 初始化一个二维数组，用于存储状态
    # dp[i][j] 表示将前 i 个物品装入容量为 j 的背包中所获得的最大价值
    dp = [[0] * (W + 1) for _ in range(N + 1)]
    # 开始进行递推
    for i in range(1, N + 1):
        for w in range(1, W + 1):
            if w - wt[i - 1] < 0:
                # 当前商品 i 的重量已经超过了 w，无法被放入当前容量为 w 的背包中，只能选择不装入背包
                dp[i][w] = dp[i - 1][w]
            else:
                # 当前商品 i 的重量小于等于当前容量 w，可以尝试将其放入背包中
                # 取最大值，考虑是将其放入之前的最优方案中还是选择不放
                dp[i][w] = max(
                    dp[i - 1][w - wt[i - 1]] + val[i - 1], #放入
                    dp[i - 1][w]  #不放入
                )
    # 返回最大价值
    return dp[N][W]
```

```python
# 416
class Solution:
    def canPartition(self, nums: List[int]) -> bool:
        s = sum(nums)
        if s%2 != 0:
            return False
        n = len(nums)
        s = s//2
        dp = [[False]*(s+1) for i in range(n+1)]
        for i in range(n+1):
            dp[i][0] = True  #如果容量为0，那么肯定能装满
        
        for i in range(1, n+1):
            for j in range(1, s+1):
                if nums[i-1] > j:  #注意这里对应的nums索引为i-1
                    dp[i][j] = dp[i-1][j]
                else:
                    dp[i][j] = dp[i-1][j] or dp[i-1][j-nums[i-1]]
        return dp[n][s]
```

```python
#518
class Solution:
    def change(self, amount: int, coins: List[int]) -> int:
        n = len(coins)
        # dp[i][j]表示给你前i个coin，凑出j金额的方法数量
        dp = [[0]*(amount+1) for i in range(n+1)]
        for i in range(n+1):
            dp[i][0] = 1
        for i in range(1, n+1):
            for j in range(1, amount+1):
                if coins[i-1] <= j: #能凑出的可能
                    dp[i][j] = dp[i-1][j] + dp[i][j-coins[i-1]]
                else:
                    dp[i][j] = dp[i-1][j]
        return dp[n][amount]
```
